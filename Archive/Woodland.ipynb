```python
# Ensure numeric types
tax_aug['Tip'] = pd.to_numeric(tax_aug['Tip'], errors='coerce')
tax_aug['Tax'] = pd.to_numeric(tax_aug['Tax'], errors='coerce')
tax_aug['Total Sum'] = pd.to_numeric(tax_aug['Total Sum'], errors='coerce')

tax_aug_summary = (
    tax_aug
      .groupby(['Order Status', 'Payment Status'], dropna=False)
      .agg(
          orders=('Order ID', 'nunique'),
          tip_sum=('Tip', 'sum'),
          tax_sum=('Tax', 'sum'),
          total_sum=('Total Sum', 'sum')
      )
      .reset_index()
      .sort_values(['Order Status', 'Payment Status'])
)

print(tax_aug_summary.head())
```

```python
# Ensure numeric types
tax_aug['Tip'] = pd.to_numeric(tax_aug['Tip'], errors='coerce')
tax_aug['Tax'] = pd.to_numeric(tax_aug['Tax'], errors='coerce')
tax_aug['Total Sum'] = pd.to_numeric(tax_aug['Total Sum'], errors='coerce')

tax_aug_summary = (
    tax_aug
      .groupby(['Order Status', 'Payment Status'], dropna=False)
      .agg(
          orders=('Order ID', 'nunique'),
          tip_sum=('Tip', 'sum'),
          tax_sum=('Tax', 'sum'),
          total_sum=('Total Sum', 'sum')
      )
      .reset_index()
      .sort_values(['Order Status', 'Payment Status'])
)

print(tax_aug_summary.head())
```

```python
# ENHANCED ANALYSIS: Detailed Missing Data and Mismatch Detection

print("=== COMPREHENSIVE ANALYSIS REPORT ===")
print(f"Total Records in trans_aug: {len(trans_aug)}")
print(f"Total Records in tax_aug: {len(tax_aug)}")
print(f"Difference: {len(trans_aug) - len(tax_aug)} records")
print()

# 1. Missing records in tax_aug (already identified)
print("=== MISSING RECORDS IN TAX_AUG ===")
missing_in_tax = trans_aug[~trans_aug['Order ID'].isin(tax_aug['Order ID'])]
print(f"Count of missing records: {len(missing_in_tax)}")

if len(missing_in_tax) > 0:
    print("\nMissing Order IDs:")
    print(missing_in_tax[['Order ID', 'Transaction Date', 'Amount', 'Location']].to_string(index=False))
    
    print(f"\nTotal amount of missing transactions: ${missing_in_tax['Amount'].sum():.2f}")
    
    # Group by location
    print("\nMissing records by location:")
    location_summary = missing_in_tax.groupby('Location').agg({
        'Order ID': 'count',
        'Amount': 'sum'
    }).rename(columns={'Order ID': 'Count', 'Amount': 'Total_Amount'})
    print(location_summary)
```

```python
# 2. MISMATCH ANALYSIS: Amount discrepancies between matching records

print("\n=== AMOUNT MISMATCH ANALYSIS ===")

# Find records that exist in both datasets
common_orders = trans_aug[trans_aug['Order ID'].isin(tax_aug['Order ID'])].copy()
common_tax_orders = tax_aug[tax_aug['Order ID'].isin(trans_aug['Order ID'])].copy()

# Merge for comparison
merged_comparison = pd.merge(
    common_orders[['Order ID', 'Amount', 'Transaction Date', 'Location']], 
    common_tax_orders[['Order ID', 'Total Sum', 'Tip', 'Tax']], 
    on='Order ID', 
    how='inner'
)

# Calculate differences
merged_comparison['Amount_Diff'] = merged_comparison['Amount'] - merged_comparison['Total Sum']
merged_comparison['Abs_Diff'] = abs(merged_comparison['Amount_Diff'])

# Find significant mismatches (difference > $0.01)
significant_mismatches = merged_comparison[merged_comparison['Abs_Diff'] > 0.01]

print(f"Total matching orders: {len(merged_comparison)}")
print(f"Orders with significant amount differences (>$0.01): {len(significant_mismatches)}")

if len(significant_mismatches) > 0:
    print("\nSignificant Amount Mismatches:")
    print(significant_mismatches[['Order ID', 'Amount', 'Total Sum', 'Amount_Diff', 'Location']].to_string(index=False))
    
    print(f"\nTotal amount discrepancy: ${merged_comparison['Amount_Diff'].sum():.2f}")
    print(f"Average absolute difference: ${merged_comparison['Abs_Diff'].mean():.2f}")
else:
    print("✅ No significant amount mismatches found!")
```

```python
# 3. DETAILED ANALYSIS OF MISSING RECORDS

print("\n=== DETAILED MISSING RECORDS ANALYSIS ===")

if len(missing_in_tax) > 0:
    print("Analyzing the 11 missing records in detail:")
    
    # Check payment types of missing records
    print("\nPayment Types of Missing Records:")
    payment_summary = missing_in_tax['Payment Type'].value_counts()
    print(payment_summary)
    
    # Check dates of missing records
    print("\nDate Distribution of Missing Records:")
    missing_in_tax['Date_Only'] = missing_in_tax['Transaction Date'].dt.date
    date_summary = missing_in_tax['Date_Only'].value_counts().sort_index()
    print(date_summary)
    
    # Check amounts
    print(f"\nAmount Range of Missing Records:")
    print(f"Min: ${missing_in_tax['Amount'].min():.2f}")
    print(f"Max: ${missing_in_tax['Amount'].max():.2f}")
    print(f"Mean: ${missing_in_tax['Amount'].mean():.2f}")
    
    # Check for zero amount transactions
    zero_amount = missing_in_tax[missing_in_tax['Amount'] == 0]
    print(f"\nZero amount transactions: {len(zero_amount)}")
    if len(zero_amount) > 0:
        print("Zero amount Order IDs:", zero_amount['Order ID'].tolist())
    
    # Check for cash transactions specifically
    cash_transactions = missing_in_tax[missing_in_tax['Payment Type'] == 'cash']
    print(f"\nCash transactions missing: {len(cash_transactions)}")
    if len(cash_transactions) > 0:
        print("Missing cash transaction Order IDs:", cash_transactions['Order ID'].tolist())
```

```python
# 4. SUMMARY REPORT AND RECOMMENDATIONS

print("\n" + "="*60)
print("FINAL SUMMARY REPORT")
print("="*60)

print(f"📊 DATASET COMPARISON:")
print(f"   • Transaction Report (trans_aug): {len(trans_aug):,} records")
print(f"   • Tax Report (tax_aug): {len(tax_aug):,} records")
print(f"   • Difference: {len(trans_aug) - len(tax_aug)} records")

print(f"\n🔍 MISSING DATA ANALYSIS:")
print(f"   • Records missing in Tax Report: {len(missing_in_tax)}")
print(f"   • Records missing in Transaction Report: 0")
print(f"   • Total missing transaction value: ${missing_in_tax['Amount'].sum():.2f}")

print(f"\n💰 AMOUNT MISMATCH ANALYSIS:")
if len(merged_comparison) > 0:
    print(f"   • Total matching records: {len(merged_comparison):,}")
    print(f"   • Records with amount differences: {len(significant_mismatches)}")
    if len(significant_mismatches) == 0:
        print("   • ✅ No significant amount mismatches found!")
    else:
        print(f"   • Total amount discrepancy: ${merged_comparison['Amount_Diff'].sum():.2f}")

print(f"\n🎯 KEY FINDINGS:")
print("   1. All missing records are in the Tax Report (not Transaction Report)")
print("   2. Missing records total $119.35 in transaction value")
print("   3. Need to investigate why these 11 transactions are not in Tax Report")

print(f"\n📋 RECOMMENDED ACTIONS:")
print("   1. Investigate the 11 missing Order IDs in the Tax Report system")
print("   2. Check if these transactions need manual tax calculation")
print("   3. Verify if cash transactions are being properly recorded in tax system")
print("   4. Review the $0.00 transaction (Order ID: 1754582417335)")

print("\n" + "="*60)
```

```python
# DETAILED LIST OF 11 MISSING RECORDS IN TAX REPORT

print("="*80)
print("DETAILED LIST OF 11 MISSING RECORDS IN TAX REPORT")
print("="*80)

if len(missing_in_tax) > 0:
    # Create a detailed table with all relevant information
    missing_details = missing_in_tax[['Order ID', 'Transaction Date', 'Amount', 'Location', 'Payment Type', 'Payment Gateway']].copy()
    
    # Sort by transaction date
    missing_details = missing_details.sort_values('Transaction Date')
    
    # Add row numbers for easy reference
    missing_details = missing_details.reset_index(drop=True)
    missing_details.index = missing_details.index + 1
    
    print("Missing Records Details:")
    print(missing_details.to_string())
    
    print(f"\nSUMMARY:")
    print(f"Total missing records: {len(missing_details)}")
    print(f"Total missing amount: ${missing_details['Amount'].sum():.2f}")
    print(f"Date range: {missing_details['Transaction Date'].min().strftime('%Y-%m-%d')} to {missing_details['Transaction Date'].max().strftime('%Y-%m-%d')}")
    
    # Payment type breakdown
    print(f"\nPayment Type Breakdown:")
    payment_breakdown = missing_details['Payment Type'].value_counts()
    for ptype, count in payment_breakdown.items():
        amount = missing_details[missing_details['Payment Type'] == ptype]['Amount'].sum()
        print(f"  {ptype}: {count} records (${amount:.2f})")
else:
    print("No missing records found.")
```

```python
# DETAILED ANALYSIS OF 4 RECORDS WITH AMOUNT DIFFERENCES

print("="*80)
print("DETAILED ANALYSIS OF 4 RECORDS WITH AMOUNT DIFFERENCES")
print("="*80)

if len(significant_mismatches) > 0:
    # Create detailed comparison table
    mismatch_details = significant_mismatches[['Order ID', 'Amount', 'Total Sum', 'Amount_Diff', 'Location', 'Transaction Date', 'Tip', 'Tax']].copy()
    
    # Sort by absolute difference (largest first)
    mismatch_details = mismatch_details.sort_values('Abs_Diff', ascending=False)
    mismatch_details = mismatch_details.reset_index(drop=True)
    mismatch_details.index = mismatch_details.index + 1
    
    print("Records with Amount Differences:")
    print(mismatch_details.to_string())
    
    print(f"\nAMOUNT DIFFERENCE ANALYSIS:")
    print(f"Total discrepancy: ${mismatch_details['Amount_Diff'].sum():.2f}")
    print(f"Average difference: ${mismatch_details['Amount_Diff'].mean():.2f}")
    print(f"Largest difference: ${mismatch_details['Amount_Diff'].max():.2f}")
    print(f"Smallest difference: ${mismatch_details['Amount_Diff'].min():.2f}")
    
    # Check if differences are related to tips or taxes
    print(f"\nTIP AND TAX ANALYSIS:")
    print(f"Total tips in tax report: ${mismatch_details['Tip'].sum():.2f}")
    print(f"Total taxes in tax report: ${mismatch_details['Tax'].sum():.2f}")
    
    # Calculate what the difference might be
    mismatch_details['Tip_Tax_Sum'] = mismatch_details['Tip'] + mismatch_details['Tax']
    mismatch_details['Diff_vs_Tip_Tax'] = mismatch_details['Amount_Diff'] + mismatch_details['Tip_Tax_Sum']
    
    print(f"\nHYPOTHESIS CHECK:")
    print("If Transaction Report = Tax Report + Tips + Taxes, then:")
    for idx, row in mismatch_details.iterrows():
        if abs(row['Diff_vs_Tip_Tax']) < 0.01:  # Very small difference
            print(f"  Order {row['Order ID']}: ✅ MATCHES (Transaction = Tax + Tips + Taxes)")
        else:
            print(f"  Order {row['Order ID']}: ❌ Still different by ${row['Diff_vs_Tip_Tax']:.2f}")
    
else:
    print("No significant amount differences found.")
```

```python
# CORRECTED ANALYSIS: Group by Order ID in Transaction Report first

print("="*80)
print("CORRECTED ANALYSIS: GROUP BY ORDER ID IN TRANSACTION REPORT")
print("="*80)

# Group transaction report by Order ID to get total amount per order
trans_aug_grouped = trans_aug.groupby('Order ID').agg({
    'Amount': 'sum',
    'Transaction Date': 'first',
    'Location': 'first',
    'Payment Type': 'first',
    'Payment Gateway': 'first'
}).reset_index()

print(f"Original trans_aug records: {len(trans_aug)}")
print(f"Grouped trans_aug records (unique Order IDs): {len(trans_aug_grouped)}")
print()

# Now compare with tax report
print("=== UPDATED MISSING RECORDS ANALYSIS ===")
missing_in_tax_corrected = trans_aug_grouped[~trans_aug_grouped['Order ID'].isin(tax_aug['Order ID'])]
print(f"Records missing in tax_aug (after grouping): {len(missing_in_tax_corrected)}")

if len(missing_in_tax_corrected) > 0:
    print("\nMissing Order IDs (grouped):")
    missing_details_corrected = missing_in_tax_corrected[['Order ID', 'Transaction Date', 'Amount', 'Location', 'Payment Type']].copy()
    missing_details_corrected = missing_details_corrected.sort_values('Transaction Date')
    missing_details_corrected = missing_details_corrected.reset_index(drop=True)
    missing_details_corrected.index = missing_details_corrected.index + 1
    print(missing_details_corrected.to_string())
    
    print(f"\nTotal missing amount (grouped): ${missing_in_tax_corrected['Amount'].sum():.2f}")
else:
    print("No missing records found after grouping.")
```

```python
# CORRECTED AMOUNT MISMATCH ANALYSIS (using grouped data)

print("\n=== UPDATED AMOUNT MISMATCH ANALYSIS ===")

# Find records that exist in both datasets (using grouped transaction data)
common_orders_corrected = trans_aug_grouped[trans_aug_grouped['Order ID'].isin(tax_aug['Order ID'])].copy()
common_tax_orders_corrected = tax_aug[tax_aug['Order ID'].isin(trans_aug_grouped['Order ID'])].copy()

# Merge for comparison
merged_comparison_corrected = pd.merge(
    common_orders_corrected[['Order ID', 'Amount', 'Transaction Date', 'Location']], 
    common_tax_orders_corrected[['Order ID', 'Total Sum', 'Tip', 'Tax']], 
    on='Order ID', 
    how='inner'
)

# Calculate differences
merged_comparison_corrected['Amount_Diff'] = merged_comparison_corrected['Amount'] - merged_comparison_corrected['Total Sum']
merged_comparison_corrected['Abs_Diff'] = abs(merged_comparison_corrected['Amount_Diff'])

# Find significant mismatches (difference > $0.01)
significant_mismatches_corrected = merged_comparison_corrected[merged_comparison_corrected['Abs_Diff'] > 0.01]

print(f"Total matching orders (grouped): {len(merged_comparison_corrected)}")
print(f"Orders with significant amount differences (>$0.01): {len(significant_mismatches_corrected)}")

if len(significant_mismatches_corrected) > 0:
    print("\nSignificant Amount Mismatches (corrected):")
    mismatch_details_corrected = significant_mismatches_corrected[['Order ID', 'Amount', 'Total Sum', 'Amount_Diff', 'Location', 'Tip', 'Tax']].copy()
    mismatch_details_corrected = mismatch_details_corrected.sort_values('Abs_Diff', ascending=False)
    mismatch_details_corrected = mismatch_details_corrected.reset_index(drop=True)
    mismatch_details_corrected.index = mismatch_details_corrected.index + 1
    print(mismatch_details_corrected.to_string())
    
    print(f"\nTotal amount discrepancy: ${merged_comparison_corrected['Amount_Diff'].sum():.2f}")
    print(f"Average absolute difference: ${merged_comparison_corrected['Abs_Diff'].mean():.2f}")
else:
    print("✅ No significant amount mismatches found after grouping!")
```
